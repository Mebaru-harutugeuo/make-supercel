```python
import numpy as np
```

```python
# unit_cell atmic distribuion 
unit_cell = np.array([
    [0, 0, 0],
    [1.75, 0, 1.75],
    [0, 1.75, 1.75],
    [1.75, 1.75, 0]
])

#　lattice constant Ni=3.542
lattice_constant = 3.5 # 任意の値に設定 Ni3.524

# 超格子の倍率Niの場合、１４３倍すると50nm
super_cell_size = np.array([30, 30, 143])

# 超格子の原子配置  
super_cell = []
for i in range(super_cell_size[0]):
    for j in range(super_cell_size[1]):
        for k in range(super_cell_size[2]):
            tranlation = np.array([i, j, k])
            for atom in unit_cell:
                super_cell.append(atom + tranlation * lattice_constant)

super_cell = np.array(super_cell)

# result
#print("超格子の原子配置:")
#for atom in super_cell:
#    print(atom)

# result .xyzファイル形式で吐き出す。
with open("atom_10.xyz", "w") as file:
    file.write(str(len(super_cell)) + "\n")  # 原子数の行を書き込み
    file.write("Generated by program\n")  # コメント行を書き込み
    for atom in super_cell:
        file.write("Ni " + " ".join(map(str, atom)) + "\n")  # 原子座標を書き込み
        
```
